```python
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
fig.suptitle("Example1:Line Plot")
ax = plt.axes()
```

```python
x = np.linspace(0,10,25)
plt.plot(x,x,'o', color='r')
plt.plot(x,x+2,'.',color='g')
plt.plot(x,x+4,'s-',color='b')
plt.plot(x,x+6,'d--',color='c')
```

```python
plt.plot(x, x+0, '-g') # solid green
plt.plot(x, x+1, '--c') # dashed cyan
plt.plot(x, x+2, '-.k') # dashdot black
plt.plot(x, x+3, ':r') # dotted red
```

```python
plt.plot(x, np.cos(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.cos(x))
plt.xlim(11, -1)
plt.ylim(1.1, -1.1)
```

```python
plt.plot(x, np.cos(x*2))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.exp(-x))
```

```python
plt.plot(x, np.exp(-x))
plt.yscale('log')
```

```python
plt.plot(x, np.cos(x))
plt.title('A Cosine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-og')
```

```python
x = np.linspace(0,10,30)
y = np.cos(x)
plt.plot(x,y,'o', color = 'gray', markersize=10, linewidth=3,
        markeredgecolor='gray', markeredgewidth=1.5)
plt.ylim(-1.2, 1.2)
```

```python
plt.scatter(x,y, marker='o')
```

```python
nrows, ncols = 3, 4

for i in range(nrows*ncols):
    plt.subplot(nrows, ncols, i+1)
    plt.text(0.5,0.5,str((nrows,ncols,i+1)), fontsize=16, ha='center')
```

```python
# 다중 서브 플롯 작성

nrows, ncols = 3,4

grid = plt.GridSpec(nrows, 4, wspace=0.4, hspace=0.4)
ax1 = plt.subplot(grid[0,0])
ax2 = plt.subplot(grid[0,1:3])
ax3 = plt.subplot(grid[1:,:2])
ax4 = plt.subplot(grid[1:,2])
ax5 = plt.subplot(grid[:,3])
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,5000).T

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4,hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1,0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1,1:], yticklabels=[],sharex=main_ax)

main_ax.plot(x,y,'ok',markersize=3, alpha=0.2)

x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y,40,histtype='stepfilled', orientation='horizontal',color='gray')
y_hist.invert_xaxis()
```
